```python
# Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
The Cumulative Distribution Function (CDF) is a function that gives the probability that a random variable X is less than or equal to a certain value x. The CDF is defined as:

F(x) = P(X ≤ x)

where X is the random variable and F(x) is the probability that X is less than or equal to x.

The CDF can be used for both discrete and continuous random variables. For discrete random variables, the CDF is a step function that increases by the probability mass at each possible value of X. For continuous random variables, the CDF is a continuous function that increases monotonically from 0 to 1 over the range of possible values of X.

For example, consider the following PMF for rolling a fair six-sided die:

x	         1	 2	 3	 4	 5	 6
P(X = x)	1/6	1/6	1/6	1/6	1/6	1/6
The CDF for this PMF is:

x	  -∞	 1	 2	 3	 4	 5	 6	+∞
F(x)	0	1/6	2/6	3/6	4/6	5/6	 1	1
This CDF indicates that the probability of rolling a number less than or equal to 3 is 3/6 = 1/2, since F(3) = 3/6. Similarly, the probability of rolling a number less than or equal to 5 is 5/6, since F(5) = 5/6.

The CDF is useful for many applications in probability and statistics. For example, the CDF can be used to calculate the probability of observing a value of X within a certain range, by taking the difference between the CDF values at the upper and lower bounds of the range. The CDF can also be used to generate random numbers from a given probability distribution, by sampling uniformly from the range [0, 1] and finding the value of X that corresponds to the CDF value closest to the sampled value.

In summary, the CDF is a useful tool for describing and analyzing probability distributions, and it provides a way to calculate probabilities and generate random numbers from a given distribution.

```

```python
# Q3: What are some examples of situations where the normal distribution might be used as a model?
# Explain how the parameters of the normal distribution relate to the shape of the distribution.

The normal distribution is a common probability distribution used to
model many different real-world situations. Some examples of 
situations where the normal distribution might be used as a model 
include:

Heights of people: The heights of people in a population tend to 
follow a normal distribution, with a mean height and a standard 
deviation that describe the average and spread of heights, 
respectively.


Errors in measurement: Errors in measurements taken in a laboratory
or manufacturing setting can often be modeled using a normal 
distribution, with the mean and standard deviation reflecting the 
accuracy and precision of the measurement process.

The parameters of the normal distribution relate to the shape of the
distribution as follows:

Mean (μ): The mean of the normal distribution determines the center 
of the distribution, or the point around which the data are 
symmetrically distributed.

Standard deviation (σ): The standard deviation of the normal
distribution determines the spread of the data, or how dispersed the
data are around the mean. Larger values of σ indicate more variability
in the data, while smaller values of σ indicate less variability.

Variance (σ^2): The variance of the normal distribution is equal to
the square of the standard deviation, and it provides a measure of 
the spread of the data in terms of the squared units of the data.

Skewness: The normal distribution has a skewness of zero, which 
means that it is symmetric around the mean. However, if the 
distribution is skewed, it indicates that the mean and median 
are not equal, and the data are not symmetrically distributed.

Overall, the normal distribution is a versatile and widely used 
probability distribution that provides a useful model for many 
different types of data. Its parameters provide important 
information about the center and spread of the data, and they can be 
used to make inferences and predictions about the data in a variety
of settings.

```

```python
# Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal
# Distribution.

The normal distribution, also known as the Gaussian distribution, is
one of the most important and widely used probability distributions 
in statistics and data analysis. It is important for several reasons:
    

1.Central Limit Theorem: The normal distribution plays a key role in
the central limit theorem, which states that the sampling distribution
of the mean of a random sample from any population with a finite mean
and variance will approach a normal distribution as the sample size 
increases. This means that even if the population from which the 
sample is drawn is not normally distributed, the distribution of
sample means will be approximately normal. This makes the normal
distribution an important tool for statistical inference and 
hypothesis testing.

2.Commonly occurring in nature: Many real-world phenomena follow a 
normal distribution, making it a useful model for understanding and
analyzing data. Some examples include the heights and weights of
people, the distribution of IQ scores, and the errors in measurements
taken in scientific experiments.

3.Ease of use and interpretation: The normal distribution is a 
mathematically simple and well-understood distribution, which makes 
it easy to calculate and interpret statistical measures such as means,
variances, and probabilities.

Examples of real-life situations that can be modeled using the
normal distribution include:

1.Heights of people: The heights of people in a population tend to 
follow a normal distribution, with a mean height and standard
deviation that describe the average and spread of heights, 
respectively.

2.IQ scores: The distribution of IQ scores in a population is 
approximately normal, with a mean of 100 and a standard deviation of 
15.

3.Errors in scientific measurements: The errors in scientific 
measurements taken in a laboratory or manufacturing setting can often
be modeled using a normal distribution, with the mean and standard
deviation reflecting the accuracy and precision of the measurement
process.

4.Stock prices: The daily returns on stock prices often follow a 
normal distribution, which is important for modeling and predicting
financial risk.

```

```python
# Q7: Explain uniform Distribution with an example.
Uniform distribution is a continuous probability distribution 
where all possible outcomes have an equal chance of occurring 
within a specified range. In other words, each value in the range
is equally likely to occur.

An example of uniform distribution is rolling a fair six-sided die.
Each face of the die has an equal probability of appearing, which is
1/6 or approximately 0.1667. Therefore, the probability density
function of a uniform distribution over the interval [a, b] is:

f(x) = 1 / (b - a) for a <= x <= b
f(x) = 0 otherwise

where a and b are the lower and upper bounds of the distribution, 
respectively.
```

```python
# Q8: What is the z score? State the importance of the z score.

In statistics, the z-score (also known as the standard score) is a
measure of how many standard deviations an observation or data point
is above or below the population mean. It is calculated by subtracting
the population mean from the observed value and dividing the result
by the population standard deviation:

z = (x - μ) / σ

where x is the observed value, μ is the population mean, and σ is the
population standard deviation.


The z-score is important because it allows us to standardize and
compare observations from different distributions with different 
scales and units. By converting the raw data to standardized units, 
we can compare observations across different samples or populations 
and make meaningful conclusions about the relative positions of the 
data points. Z-scores are used in hypothesis testing, confidence 
intervals, and in constructing control charts
```

```python
# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) states that if you take a large number of random samples from any population 
with any distribution, the distribution of the sample means will
follow a normal distribution. This means that even if the population 
itself does not have a normal distribution, the average of many 
samples will tend to have a normal distribution. The CLT is important
because it allows us to use statistical methods that rely on the 
normal distribution, even if we dont know the 
population distribution, as long as we have a large enough sample
size.

The significance of the Central Limit Theorem is that it allows us to
make inferences about the population mean from a sample mean, even if
the population distribution is unknown or non-normal. It forms the
basis of many statistical techniques, including hypothesis testing, 
confidence intervals, and estimation.
```

```python
# Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem has a few key assumptions:

1.Random Sampling: The sample must be randomly selected from the
population.

2.Sample Size: The sample size must be large enough. There is no
hard and fast rule for what constitutes a "large" sample, but a
general rule of thumb is that the sample size should be greater
than or equal to 30.

3.Independence: The observations in the sample must be independent of
each other.

4.Finite Variance: The population distribution from which the samples
are drawn must have a finite variance.

It is important to note that violating any of these assumptions can
result in the Central Limit Theorem not being applicable.
```
