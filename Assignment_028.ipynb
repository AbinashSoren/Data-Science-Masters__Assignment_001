```python
# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation statistics is a branch of statistics that involves 
estimating unknown population parameters from sample data. There are 
two types of estimation: point estimation and interval estimation.

Point estimation involves using a single value (a point) to estimate
an unknown population parameter. For example, if we want to estimate
the population mean, we can use the sample mean as a point estimate. 
Point estimates are simple to calculate and easy to understand, but 
they do not provide any information about the variability of the 
estimate.

Interval estimation involves using a range of values (an interval) to
estimate an unknown population parameter. For example, we can use a 
confidence interval to estimate the population mean. A confidence
interval is a range of values that is likely to contain the true
population parameter with a specified degree of confidence. 
Confidence intervals provide information about the variability of
the estimate and can give a more accurate picture of the population 
parameter.

Interval estimation is typically preferred over point estimation
because it provides more information about the variability of the
estimate and is therefore more informative. However, it is also more
complex to calculate and interpret than point estimation.




```

```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard
# deviation.

import math

def estimate_mean(sample_mean, sample_std, sample_size):
    # Calculate the standard error of the mean
    std_error = sample_std / math.sqrt(sample_size)
    
    # Calculate the margin of error for a 95% confidence interval
    margin_error = 1.96 * std_error
    
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    
    # Return the estimated population mean and the confidence interval
    return (sample_mean, lower_bound, upper_bound)

sample_mean = 10.5
sample_std = 3.2
sample_size = 100

estimated_mean, lower_bound, upper_bound = estimate_mean(sample_mean, sample_std, sample_size)

print("Estimated Population Mean:", estimated_mean)
print("95% Confidence Interval:", lower_bound, "-", upper_bound)
```

```python
# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

Hypothesis testing is a statistical method used to make decisions 
about a population based on a sample of data. It involves testing a
hypothesis (a statement or assumption about the population) by
examining the sample data and drawing a conclusion about whether the
hypothesis is supported or not.

Hypothesis testing is used to determine whether there is enough 
evidence to support a claim about a population. It can be used to 
test a wide range of hypotheses, such as whether the mean of a 
population is equal to a certain value, whether two populations have
the same mean, or whether a new treatment is effective.

The importance of hypothesis testing lies in its ability to provide a
rigorous and objective way to make decisions based on data. By using 
hypothesis testing, we can avoid making decisions based on intuition 
or personal bias, and instead rely on statistical evidence to support
our conclusions. This is particularly important in scientific research
and in making business decisions, where decisions can have significant
consequences.

Hypothesis testing involves setting up a null hypothesis (H0), which 
represents the status quo or the assumption that there is no
difference or no effect. The alternative hypothesis (Ha) represents 
the claim or the hypothesis that we want to test. The sample data is 
then used to calculate a test statistic, which is compared to a 
critical value or p-value to determine whether the null hypothesis
should be rejected or not. If the p-value is less than the 
significance level (usually set at 0.05), we reject the null
hypothesis and conclude that there is enough evidence to support
the alternative hypothesis. If the p-value is greater than the
significance level, we fail to reject the null hypothesis and 
conclude that there is not enough evidence to support the alternative
hypothesis.
```

```python
# Q5.Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.

import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples from two populations
np.random.seed(123)
sample1 = np.random.normal(10, 2, 100)
sample2 = np.random.normal(12, 2, 100)

# print(sample1,sample2)
# Compute the sample means and standard deviations
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Conduct the two-sample t-test
t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)

# Print the results
print("Sample 1 Mean: {:.2f}".format(mean1))
print("Sample 1 Standard Deviation: {:.2f}".format(std1))
print("Sample 2 Mean: {:.2f}".format(mean2))
print("Sample 2 Standard Deviation: {:.2f}".format(std2))
print("T-statistic: {:.2f}".format(t_stat))
print("P-value: {:.4f}".format(p_value))
```

```python
# Q6: What is a null and alternative hypothesis? Give some examples.

In hypothesis testing, the null hypothesis (H0) is a statement of 
"no effect" or "no difference" between the two populations being
compared, while the alternative hypothesis (Ha) is a statement that 
there is some difference or effect between the two populations.

Some examples of null and alternative hypotheses are:

Null hypothesis: The mean height of male and female students in a
school is the same. Alternative hypothesis: The mean height of male
and female students in a school is different.

Null hypothesis: A new drug does not improve the recovery time for
patients compared to the standard treatment. Alternative hypothesis:
The new drug reduces the recovery time for patients compared to the 
standard treatment.

Null hypothesis: The proportion of defective products produced by a 
machine is equal to or less than 10%. Alternative hypothesis: The 
proportion of defective products produced by a machine is greater
than 10%.

The null and alternative hypotheses are typically stated before
collecting any data, and the hypothesis test is conducted to either
reject or fail to reject the null hypothesis based on the evidence in
the data.



```

```python
# Q7: Write down the steps involved in hypothesis testing.

The following are the general steps involved in hypothesis testing:

1.State the null and alternative hypotheses: The first step is to 
clearly state the null and alternative hypotheses. The null hypothesis
usually assumes that there is no significant difference between the 
sample and the population, while the alternative hypothesis is what
the researcher wants to show.

2.Set the level of significance: The level of significance (alpha) is 
the probability of rejecting the null hypothesis when it is actually 
true. The commonly used level of significance is 0.05.

3.Choose the appropriate test statistic: The choice of the appropriate
test statistic depends on the type of data and the hypothesis being 
tested. Some commonly used test statistics include t-test, z-test, 
chi-square test, ANOVA, etc.

4.Determine the critical value or p-value: Based on the chosen test 
statistic and level of significance, the critical value or p-value is
calculated. The critical value is compared to the test statistic to
determine whether to reject or fail to reject the null hypothesis, 
while the p-value represents the probability of obtaining the observed
test statistic or a more extreme one, assuming the null hypothesis is
true.

5.Collect the data and calculate the test statistic: Once the null
and alternative hypotheses are stated, the data is collected, and the
test statistic is calculated using the appropriate formula.

6.Make a decision: Based on the critical value or p-value calculated 
in step 4, a decision is made to either reject or fail to reject the 
null hypothesis.

7.Draw conclusions: The final step is to draw conclusions based on the 
results of the hypothesis test. If the null hypothesis is rejected,
the alternative hypothesis is accepted, and it can be concluded that 
there is evidence of a significant difference between the sample and 
the population. If the null hypothesis is not rejected, it can be 
concluded that there is not enough evidence to support the 
alternative hypothesis.
```

```python
# Q8. Define p-value and explain its significance in hypothesis testing.

In hypothesis testing, the p-value is the probability of obtaining
the observed test statistic or a more extreme one, assuming the null
hypothesis is true. It represents the level of evidence against the 
null hypothesis, with a smaller p-value indicating stronger evidence 
against the null hypothesis.

The significance of the p-value is that it allows researchers to 
make an objective decision about whether to reject or fail to reject
the null hypothesis. If the p-value is less than or equal to the 
level of significance (usually 0.05), the null hypothesis is rejected,
and the alternative hypothesis is accepted. This means that there is 
sufficient evidence to support the alternative hypothesis, and the
observed difference in the sample is unlikely to have occurred by 
chance.

On the other hand, if the p-value is greater than the level of 
significance, the null hypothesis is not rejected, and it can be 
concluded that there is not enough evidence to support the alternative
hypothesis. This does not necessarily mean that the null hypothesis 
is true, but rather that there is not enough evidence to reject it.

In summary, the p-value is a critical component of hypothesis testing
because it provides a measure of the strength of evidence against the 
null hypothesis and helps researchers make informed decisions about 
whether to accept or reject the alternative hypothesis.



```

```python
# Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
# parameter set to 10.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10  # degrees of freedom
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 100)  # generate 100 points on the x-axis
pdf = t.pdf(x, df)  # calculate the probability density function

# plot the probability density function
plt.plot(x, pdf, label='t-distribution')

plt.xlabel('x')
plt.ylabel('Probability density')
plt.title('Student\'s t-distribution (df=10)')
plt.legend()
plt.show()


This code generates a plot of the Students t-distribution with 10
degrees of freedom. The t.ppf() function is used to generate 100 
points on the x-axis, and the t.pdf() function is used to calculate
the probability density function at each point. The resulting plot
shows the probability density function of the t-distribution, which 
has a peak at the mean (0 in this case) and thicker tails than the 
normal distribution.
```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.

import numpy as np
from scipy.stats import t

# Generate two random samples of equal size
sample1 = np.random.normal(loc=10, scale=2, size=20)
sample2 = np.random.normal(loc=12, scale=3, size=20)

# Calculate the sample means and standard deviations
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the degrees of freedom
df = len(sample1) + len(sample2) - 2

# Calculate the pooled standard deviation
sp = np.sqrt(((len(sample1)-1)*std1**2 + (len(sample2)-1)*std2**2) / df)

# Calculate the t-statistic
t_statistic = (mean1 - mean2) / (sp * np.sqrt(1/len(sample1) + 1/len(sample2)))

# Calculate the p-value
p_value = 2 * (1 - t.cdf(np.abs(t_statistic), df))

# Print the results
print(f'Sample 1 mean: {mean1:.2f}')
print(f'Sample 2 mean: {mean2:.2f}')
print(f'Sample 1 standard deviation: {std1:.2f}')
print(f'Sample 2 standard deviation: {std2:.2f}')
print(f'Degrees of freedom: {df}')
print(f'Pooled standard deviation: {sp:.2f}')
print(f'T-statistic: {t_statistic:.2f}')
print(f'P-value: {p_value:.4f}')


# In this case, the p-value is 0.0688, which is greater than the 
# significance level of 0.05. Therefore, we fail to reject the null 
# hypothesis that the population means are equal, at a 5% level of 
# significance
```

```python
# Q12: What is t-statistic? State the formula for t-statistic.

t-statistic is a measure of the difference between the sample mean 
and the hypothesized population mean, in units of the standard error 
of the sample mean. It is used in hypothesis testing to determine 
whether the difference between two sample means is statistically 
significant or just due to chance.

The formula for t-statistic is:

t = (x̄1 - x̄2) / (s√(1/n1 + 1/n2))

where:

x̄1 and x̄2 are the sample means of two independent samples
s is the pooled standard deviation of the two samples
n1 and n2 are the sample sizes of the two independent samples
The t-statistic follows a Student's t-distribution with (n1+n2-2)
degrees of freedom.
```

```python
# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
# clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
# standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

To test the hypothesis, we need to set up the null and 
alternative hypotheses:

Null hypothesis (H0): The new drug will not decrease blood pressure 
by 10 mmHg or more.

Alternative hypothesis (Ha): The new drug will decrease blood
pressure by more than 10 mmHg.

We can use a one-sample t-test to test the hypothesis.
The t-statistic is calculated as:

t = (sample mean - hypothesized mean) / (standard deviation / sqrt(sample size))

where sample mean = 8, hypothesized mean = 10, standard deviation = 3,
and sample size = 100.

Substituting the values, we get:

t = (8 - 10) / (3 / sqrt(100)) = -6.67

Using a t-table with 99 degrees of freedom (n-1), at a significance
level of 0.05, we find the critical value to be -1.660. Since our 
calculated t-value (-6.67) is less than the critical value, we reject
the null hypothesis.

Therefore, we can conclude that the new drug does decrease 
blood pressure by more than 10 mmHg.




```

```python
# Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
# standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
# is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
# pounds with a significance level of 0.01.


We can use a one-tailed t-test to test this hypothesis. The null 
hypothesis is that the true mean weight of the products is equal to 
5 pounds, and the alternative hypothesis is that the true mean weight 
is less than 5 pounds.

The test statistic can be calculated as follows:

t = (sample_mean - population_mean) / (standard_error)

where sample_mean = 4.8, population_mean = 5, 
standard_error = standard_deviation / sqrt(sample_size) = 0.5 / sqrt(25) = 0.1.

Plugging in the values, we get:

t = (4.8 - 5) / 0.1 = -2

The degrees of freedom for this test is 24 (sample_size - 1).

Using a t-table or a calculator, we find the critical t-value for
a one-tailed test with a significance level of 0.01 and 24 degrees of
freedom to be -2.492.

Since our calculated t-value of -2 is greater than the critical
t-value of -2.492, we reject the null hypothesis and conclude that
the true mean weight of the products is less than 5 pounds with a 
significance level of 0.01.
```

```python
# Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
# 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
# score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
# groups are equal with a significance level of 0.01.


To test the hypothesis that the population means for the two groups
are equal, we can use a two-sample t-test for independent samples.
The null hypothesis is that the population means are equal, and the 
alternative hypothesis is that they are not equal.

The test statistic for the two-sample t-test is calculated as:

t = (x1 - x2) / sqrt(s1^2/n1 + s2^2/n2)

where x1 and x2 are the sample means, s1 and s2 are the sample 
standard deviations, and n1 and n2 are the sample sizes.

Using the given values, we can calculate the test statistic as:

t = (80 - 75) / sqrt(10^2/30 + 8^2/40) = 1.61

The degrees of freedom for the test are calculated as:

df = (s1^2/n1 + s2^2/n2)^2 / ((s1^2/n1)^2/(n1-1) + (s2^2/n2)^2/(n2-1))

df = (10^2/30 + 8^2/40)^2 / ((10^2/30)^2/29 + (8^2/40)^2/39) = 64.92 (rounded to 2 decimal places)

Using a t-table or a t-distribution calculator with 64 degrees of
freedom and a significance level of 0.01, we can find the critical
t-value to be approximately 2.39 (rounded to 2 decimal places).

Since the calculated t-value of 1.61 is less than the critical
t-value of 2.39, we fail to reject the null hypothesis. Therefore,
we do not have enough evidence to conclude that the population means 
for the two groups are different at a significance level of 0.01.
```

```python
# Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV
# program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard
# deviation of 1.5. Estimate the population mean with a 99% confidence interval.

To estimate the population mean with a 99% confidence interval, we can use the following formula:

Confidence interval = sample mean ± (t-value)*(standard error)

Where:

Sample mean = 4
Standard deviation = 1.5
Sample size = 50
t-value for a 99% confidence level with 49 degrees of freedom (n-1) is 2.68 (can be obtained from t-distribution table or using a function in Python)
First, we need to calculate the standard error, which is:

Standard error = standard deviation / sqrt(sample size)

Standard error = 1.5 / sqrt(50) = 0.2121

Substituting the values in the formula, we get:

Confidence interval = 4 ± (2.68)*(0.2121)

Confidence interval = 4 ± 0.568

Therefore, the 99% confidence interval for the population mean number
of ads watched by viewers during a TV program is (3.432, 4.568). 
We can be 99% confident that the true population mean falls within 
this interval.



```
