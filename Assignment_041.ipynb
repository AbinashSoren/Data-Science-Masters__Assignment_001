```python
# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
# distribution.

import pandas as pd
import matplotlib.pyplot as plt

# Load the flight dataset into a pandas DataFrame
df = pd.read_excel('flight_price.xlsx')

# Extract the "Price" column from the DataFrame
prices = df["Price"]

# Create a histogram of the "Price" column
plt.hist(prices, bins=20)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Flight Prices")
plt.show()
```

```python
# Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

import pandas as pd
import matplotlib.pyplot as plt

# Load the flight dataset into a pandas DataFrame
df = pd.read_excel('flight_price.xlsx')

# Extract the "Price" column from the DataFrame
price = df["Price"]

print('The max price of the flight is',max(price))
print('The min price of the flight is',min(price))
print('The range for price of the flight is' ,max(price)-min(price))
```

```python
# Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
# airlines.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flight dataset into a pandas DataFrame
df = pd.read_excel('flight_price.xlsx')

# Extract the "Price" and "Airline" columns from the DataFrame
prices_by_airline = df[["Price", "Airline"]]
fig, _ = plt.subplots(figsize=(15, 10))

print(prices_by_airline.info())
# Create a boxplot of the "Price" column, with the "Airline" column as the grouping variable
sns.boxplot(x="Airline", y="Price", data=prices_by_airline)
plt.xticks(rotation=30)
plt.title("Price Distribution by Airline")
plt.show()
```

```python
# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
# the relationship.

import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Remove rows with missing values in the Rating and Size columns
df = df.dropna(subset=['Rating', 'Size'])


df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].str.replace('+','')
df['Size']=df['Size'].str.replace(',','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
df['Size']=df['Size'].astype(float)
fig, _ = plt.subplots(figsize=(20, 15))
# Create a scatter plot
sns.scatterplot(data=df, x='Size', y='Rating')

# Show the plot
plt.show()
```

```python
# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Convert the Price column to numeric values
df['Price'] = df['Price'].apply(lambda x: float(x[1:]) if x[0] == '$' else 0)

# Group the data by Type and calculate the mean price for each type
# mean_prices = df.groupby('Type')['Price']
df['Price']=df['Price'].astype(float)
# Create a bar chart
num_free = df[df['Price'] == 0].shape[0]
num_paid = df[df['Price'] > 0].shape[0]

# Create a bar chart
plt.bar(['Free', 'Paid'], [num_free, num_paid])

# Add chart labels and title
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.title('Average App Price by Type')

# Show the plot
plt.show()
```

```python
# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
# with the highest number of installs.

import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Remove any rows with missing values
df = df.dropna()

# Convert the Installs column to numeric values
df['Installs'] = df['Installs'].apply(lambda x: int(x[:-1].replace(',', '')))

# Group the data by app name and calculate the total number of installs
install_counts = df.groupby('App')['Installs'].sum()

# Sort the install counts in descending order and take the top 10
top_10_apps = install_counts.sort_values(ascending=False)[:10]

# Create a frequency table for the top 10 apps
freq_table = pd.DataFrame({'App': top_10_apps.index, 'Total Installs': top_10_apps.values})

# Display the frequency table
print(freq_table)
```
