```python
Q5. What are Custom Exceptions in python? Why
do we need Custom Exceptions? Explain with
an example ?

Custom Exception are exception created by the user as per his/her's
need to tacle a specic type of error.
We use Custom Exception because it enhances the ease in 
debugging and we get customized error messages so that 
the debgging becomes easily traceble.

class InvalidAgeException(Exception):
    def __init__(self):
        print('Hi..I am from InvalidAgeException')

# you need to guess this number
number = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
        
except InvalidAgeException :
    print("Exception occurred: Invalid Age")
```
