```python
Q2: Why threading module used? Write the use of the following functions
1.activeCount():Returns the Number of Thread objects currently actve.
2.currentThread():Refers to the current Thread Running.
3.enumerate():Returns a list of Thread that are currently active.

Example:
import time
from threading import *

def thread_1(i):
    time.sleep(3)
    print("Number of active threads:",active_count())
    print('Value by Thread 1:', i)
    print('The Thread Currently is running is :',current_thread().name)


def thread_2(i):
    time.sleep(1)
    print("Number of active threads:",active_count())
    print('Value by Thread 2:', i)
    print('The Thread Currently is running is :',current_thread().name)
    
    
    
def thread_3(i):
    time.sleep(5)
    print("Number of active threads:",active_count())
    print("Value by Thread 3:", i)
    print('The Thread Currently is running is :',current_thread().name)


# Creating sample threads 
thread1 =Thread(target=thread_1, args=(100,),name='ThisIsThreadNumber1')
thread2 =Thread(target=thread_2, args=(200,),name='ThisIsThreadNumber2')
thread3 =Thread(target=thread_3, args=(300,),name='ThisIsThreadNumber3')

print("Number of active threads in the starting:",active_count())
print("The active threads in the starting is one which is the main thread that executes till the program runs")
print('The List of Thread before starting the child Thread are :',enumerate())
# Starting the threads
thread1.start()
thread2.start()
thread3.start()
print('Is Thread_1 alive',thread1.is_alive())
print('Is Thread_2 alive',thread2.is_alive())
print('Is Thread_3 alive',thread3.is_alive())
thread1.join()
thread2.join()
thread3.join()
print('The List of Thread after starting the child Thread are :',enumerate())

```

```python
# Q4. Write a python program to create two threads. Thread one must print the list of squares and thread
# two must print the list of cubes.

from threading import*

def list_of_square(temp_list):
    for x in temp_list:
        print('The Square of number {} is {}'.format(x,x**2))

def list_of_cube(temp_list):
    for x in temp_list:
        print('The Cube of number {} is {}'.format(x,x**3))
        
temp_list=[i for i in range(0,11)]        
t1=Thread(target=list_of_square,args=(temp_list,),name='Thread_Executing_fn_list_of_square')        

t2=Thread(target=list_of_cube,args=(temp_list,),name='Thread_Executing_fn_list_of_cube')

t1.start()
t2.start()
```

```python
Q6. Explain deadlocks and race conditions.

A deadlock is concurrency failure mode where a thread or
threads wait for condition that never occurs.

One of the Example can be that a Thread tries to attempts
to acquire the same lock twice.

Code:
from threading import *
def task(lock):
    print('Thread acquring Lock')
    with lock:
        print('Thread Trying to acquire lock again')
        with lock:
            pass
lock=Lock()
thread_1=Thread(target=task,args=(lock,),name='Mera_Wala_Thread')
thread_1.start()
thread_1.join()  

A race condition is a bug in concurrency programming.

It is a failure case where the behavior of the program is
dependent upon the order of execution by two or more threads.
This means, the behavior of the program will not be 
predictable, possibly changing each time it is run.

Code Example:
from threading import *
import time
x=10

def increment(increment_by):
    global x
    
    local_counter=x
    local_counter+=increment_by
    
    time.sleep(1)
    x=local_counter
    print(f'{current_thread().name} increments x by {increment_by} x: {x}')

t1=Thread(target=increment,args=(5,),name='mera_thread_1')   

t2=Thread(target=increment,args=(10,),name='mera_thread_2')

t1.start()
t2.start()

t1.join()
t2.join()

print(f'The final value of x is {x}')
```
