```python
# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

Class:
    A class contains the blueprints or the prototype from which the objects are being created.
    It is a logical entity that contains some attributes and methods. 
    
Object:
    The object is an entity that has a state and behavior associated with it.

class Human:
    
    isHuman=True #Class Attribute
    def __init__(self,name,age,skills): #constructor
        self.name=name                  #instance_variable_1              
        self.age=age                    #instance_variable_2 
        self.skills=skills              #instance_variable_3
    
    def details(self):
        print('Hi..There my name is {}.Nice to meet you.!'.format(self.name))
        print('I am {} years old.'.format(self.age))
        print('Currently i acquire the following skills')
        for x in self.skills:
            print('\t->{}'.format(x))
person1=Human('Abinash',25,['Python','Machine Learning','Deep Learning','NLP'])    #Creating a new object of type class Human
person1.details()
```

```python
# Q2. Name the four pillars of OOPs.

The four pillar of OOPS:
    ->Inheritance:  The ability of the child class to inherit or derive  the properties parent class.
    ->Polymorphism: Polymorphism refers to having multiple forms. Polymorphism is a programming term that refers to the use of the same function name, but with different signatures, for multiple types.
    ->Encapsulation: It describes the idea of wrapping data and the methods that work on data within one unit. This puts restrictions on accessing variables and methods directly and can prevent the accidental modification of data.
    ->Abstraction:Abstraction is used to hide the internal functionality of the function from the users. 
```

```python
# Q5. What is inheritance? Give an example for each type of inheritance.

Inheritance is the ability of the child class to inherit properties
from parent class.

There are 5 types of Inheritance in Python:
    1.>Single Inheritance: Single Inheritance is enabled when child class 
    inherits properties from a single parent.
    
    Like:
            A(Parent class)
            |
            |
            V
            B(Child class)
    Code:
class A: #(Parent class)
    var_1=10
    var_2=20
    def __init__(self):
        print('In class A')
    class B(A):  #class B inheriting the class A so it is child class
        pass
        
    obj=B()    #since B does not have constructor it will call the contructor for parent class (A)
    print(obj.var_1)  #able to access the variable defined in Parent class
    print(obj.var_2)
    
2.Multiple Inheritance:When the class can be derieved from more
than one base class.
    
    Like:
            A(Parent class_1)   B(Parent class_2)
                  \                  /
                   \                /
                    \              / 
                     \            / 
                      \          /
                       \        /  
                        \      / 
                         \    / 
                          \  /
                     C(Child Class)
    Code:
class A:#(Parent class)
    var_1=10
    var_2=20
    def display_1(self):
        print('Hello from class A')
class B():  
    var_3=30
    var_4=40
    def display_2(self):
        print('Hi This is class B')
class C(A,B):#class C inheriting the class A and class B ,so class C is child class
    pass
        
obj=C()    
#accessing data and method from class A
print(obj.var_1)  
print(obj.var_2)
obj.display_1()
print('*---'*10)
#accessing data and method from class B
print(obj.var_3)  
print(obj.var_4)
obj.display_2()

3.MulitLevel Inheritance:
    In multilevel inheritance, the intermediatory class inherits
    from base class, and the child class inherits from intermediatory
    class.
    
    Like:
         Like:
            A(Parent class)
            |
            |
            V
            B(Intermediatory class)
            |
            |
            V
            C(Child Class)
# Code:
class A: #(Parent class)
    var_1=10
    var_2=20
    def display_1(self):
        print('In Parent class')
class B(A):  
    var_3=30
    var_4=40
    def display_2(self):
        print('In Intermediatory class ')
class C(B):
    pass
obj=C()    
#accessing data and method from Parent class A
print(obj.var_1)  
print(obj.var_2)
obj.display_1()
print('*---'*10)
#accessing data and method from Intermidiatory class B
print(obj.var_3)  
print(obj.var_4)
obj.display_2() 

4.Hierarchical Inheritance:
    When more than one derived class are created from a single base this 
    type of inheritance is called hierarchical inheritance.
    
    Like:
        
        -----------A-----------  #Parent
        |          |          |
        B          C          D
    (child_1)   (child_2)  (child_3)
    
    Code:
class A: #(Parent class)
    var_1=10
    var_2=20
    def __init__(self):
        print('In class A')
class B(A):  #class B inheriting the class A so it is child_1 class
    pass
        
class C(A):  #class C inheriting the class A so it is child_2 class
    pass
        
class D(A):  #class D inheriting the class A so it is child_3 class
    pass
        
obj=B()    #since B does not have constructor it will call the contructor for parent class (A)
print(obj.var_1)  #able to access the variable defined in Parent class
print(obj.var_2) 

obj=C()    #since C does not have constructor it will call the contructor for parent class (A)
print(obj.var_1)  #able to access the variable defined in Parent class
print(obj.var_2)

obj=D()    #since D does not have constructor it will call the contructor for parent class (A)
print(obj.var_1)  #able to access the variable defined in Parent class
print(obj.var_2)

5:Hybrid Inheritance:Inheritance consisting of multiple types of inheritance
is called hybrid inheritance.
Code:
class A:
    def display_1(self):
        print("Inside Class A")

class B(A):
    def display_2(self):
        print("Inside Class B")

class C(A):
    def display_3(self):
        print("Inside Class C")

class D(B, C):
    def display_4(self):
        print("Inside Class D")

# create object
obj = D()

obj.display_1()
obj.display_2()
obj.display_3()
obj.display_4()
```
