```python
# Q1. What is Abstraction in OOps? Explain with an example.

Abstraction is one of the main pillars of OOPs(Object Oriented Programming).
With, Abstraction we hide the unnescessary code details from the user i.e
we do not want to give sensitive parts of our code implementaion.

Example:
from abc import ABC
class Mammals(ABC):
    def details(self):
        pass

class HumanBeing(Mammals):
    def details(self):
        print('Hi..I am a Human Being and also a Mammal')
class Animal(Mammals):
    def details(self):
        print('Hi..I am an Animal and also a Mammal')

human_being_1=HumanBeing()
human_being_1.details()

animal_1=Animal()
animal_1.details()
```

```python
# Q2. Differentiate between Abstraction and Encapsulation. Explain with an example.

Data Abstraction can be described as the technique of hiding internal details of a program and exposing the functionality only.
Implementation hiding is done using this technique.

Data Encapsulation can be described as the technique of binding up of data along with its correlate methods as a single unit.
Information hiding is done using this technique.

#Abstarction

from abc import ABC,abstractmethod
class Shapes(ABC):
    @abstractmethod
    def details(self):
        pass
    @abstractmethod
    def area(self):
        pass

class Circle(Shapes):
    def details(self):
        print('I am from Circle Class')
    def area(self):
        radius=int(input('Enter radius'))
        print('Area of Circle with radius {} is {}'.format(radius,3.14*radius**2))
class Rectangle(Shapes):
    def details(self):
        print('I am from Rectangle Class')
    def area(self):
        leng=int(input('Enter Length'))
        bre=int(input('Enter Breadth'))
        print('Area of the Reactangle with Length as {} and Breadth as {} is {}'.format(leng,bre,leng*bre))
cir=Circle()
cir.details()
cir.area()
print('*--'*30)
rec=Rectangle()
rec.details()
rec.area()

#Encapsulation
class Circle():
    def __init__(self):
        self.__rad=0
    
    def get_rad(self):
        return self.__rad
    
    def set_rad(self,rad):
        if type(rad)==int:
            self.__rad=rad
    def area(self):
        print('Area of Circle  with radius {} is {}'.format(self.get_rad(),3.14*self.get_rad()**2))

c=Circle()
print(c.get_rad())
c.area()
print('*---'*25)
c.set_rad(5)
print(c.get_rad())
c.area()
```

```python
Q3. What is abc module in python? Why is it used?

The 'abc' module in Python library provides the infrastructure 
for defining custom abstract base classes.

'abc' works by marking methods of the base class as abstract. 
This is done by @absttractmethod decorator. 
A concrete class which is a sub class of such abstract base class
then implements the abstract base by overriding its abstract methods.
```

```python
Q4. How can we achieve data abstraction?

Python doesnt provide the abstract class itself. We need to import the
abc module, which provides the base for defining Abstract Base classes
(ABC). The ABC works by decorating methods of the base class as
abstract. It registers concrete classes as the implementation of the 
abstract base. We use the @abstractmethod decorator to define an 
abstract method or if we dont provide the definition to the method, 
it automatically becomes the abstract method.

Code Example:
from abc import ABC,abstractmethod
class Shapes(ABC):
    @abstractmethod
    def details(self):
        pass
    @abstractmethod
    def area(self):
        pass
```
